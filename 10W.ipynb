```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
w10_flow=[]
w10_density=[]

mag_txt_read=open("10w_flow.txt","r")
read_list=mag_txt_read.readlines()
for item in read_list:
    w10_flow.append(float(item))
    
mag_txt_read=open("10w_density.txt","r")
read_list=mag_txt_read.readlines()
for item in read_list:
    w10_density.append(float(item))
```

```python
print(len(w10_flow))
print(len(w10_density))
```

```python
pflow=[]
pdensity=[]
loopi=12



i=0

while i<len(w10_flow)-5:
    
    sum1=0
    for j in range(i,i+loopi):
        sum1+=w10_flow[j]
    
    pflow.append(sum1/loopi)
    
    i+=loopi

    
    
i=0

while i<len(w10_density)-5:
    
    sum1=0
    for j in range(i,i+loopi):
        sum1+=w10_density[j]
    
    pdensity.append(sum1/loopi)
    
    i+=loopi
    
    
print(len(pflow))
print(len(pdensity))
```

```python
i=0
spflow=[]
spdensity=[]
colorl=[]
while i<len(pflow):
    spflow.append(pflow[i])
    spdensity.append(pdensity[i])
    colorl.append(i*23/288)
    i=i+2
```

```python
plt.figure(figsize=(5,5))
plt.plot(spdensity, spflow, 'o', color='black');
```

```python
plt.figure(figsize=(8,6.5))
plt.scatter(spdensity, spflow, c=colorl, cmap="RdYlGn", s=20, edgecolors="black")
plt.colorbar()
plt.ylabel('flow(vehicle per 5 minutes)')
plt.xlabel('occupancy')
```
